```python
def transpose(matrix):
    transposed=np.zeros((len(matrix[0]), len(matrix)))
    for i in range(len(matrix[0])):
        for j in range(len(matrix)):
            transposed[i][j]=matrix[j][i]
    return transposed
            
```

```python
def matmul(x, y):
    x=np.asarray(x)
    y=np.asarray(y)
    if len(x[0])==len(y):
        result=np.zeros((len(x), len(y[0])))
        return x@y
    else:
        return None
    
    
```

```python
def relu(matrix):
    return np.maximum(0, matrix)
```
